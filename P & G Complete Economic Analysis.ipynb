## Production Analysis

```python
columns_production_analysis_data=["Year", "Sales", "No. of Employees", "Total Capital"]
```

```python
production_analysis_data=pd.DataFrame(list(zip(Full_data["Year"][8:],Full_data["Sales"][8:],Full_data["No. Of employees"][8:],
                                        Full_data["Total Capital"][8:])), columns= columns_production_analysis_data)
```

```python
production_analysis_data["Sales"]=production_analysis_data["Sales"]*1000000
production_analysis_data["Total Capital"]=production_analysis_data["Total Capital"]*1000000
```

```python
pd.set_option('display.float_format', lambda x: '%.f' % x) #so numbers don't appear in scientific notation
```

```python
production_analysis_data['No. of Employees']=production_analysis_data['No. of Employees'].apply(
    lambda x: int(x) if math.isnan(x) == False else "N/A")
```

```python
production_analysis_data
```

#### Log function

```python
def log_of_columns(df, list_of_columns):
    for i in list_of_columns:
        df["Log of " + df.columns[i]]=df[df.columns[i]].apply(lambda x: math.log(x) if isinstance(x, str) == False else "N/A")
        df["Log of " + df.columns[i]]=df["Log of " + df.columns[i]].apply(lambda x: '%.5f' % x if isinstance(x, str) == False else "N/A")
        
```

```python
log_columns=[1,2,3]
log_of_columns(production_analysis_data, log_columns)
```
