```python
byr (Birth Year) - four digits; at least 1920 and at most 2002.
iyr (Issue Year) - four digits; at least 2010 and at most 2020.
eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
hgt (Height) - a number followed by either cm or in:
If cm, the number must be at least 150 and at most 193.
If in, the number must be at least 59 and at most 76.
hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
pid (Passport ID) - a nine-digit number, including leading zeroes.
cid (Country ID) - ignored, missing or not.
```

```python
def valid_or_not(dic):
    if int(dic['byr'])>= 1920 and int(dic['byr'])<= 2002:
        valid = True
    else:
        valid = False
        return valid
    
    if int(dic['iyr'])>=2010 and int(dic['iyr'])<= 2020:
        valid = True
    else:
        valid = False
        return valid
    
    if int(dic['eyr'])>= 2020 and int(dic['eyr'])<= 2030:
        valid = True
    else:
        valid = False
        return valid
   
    ht=dic['hgt']
    if len(ht) >= 4:
        ht_n=int(ht[:-2])
        if ht.endswith('cm') and ht_n>=150 and ht_n<=193:
            valid = True
        elif ht.endswith('in') and ht_n>=59 and ht_n<=76:
            valid= True
        else:
            valid= False
            return valid
    else:
        valid=False
        return valid
    
    color=dic['hcl']
    if len(color) != 7 and color[0] != '#' or any([x not in '0123456789abcdef' for x in color[1:]]):
        valid = False
        return valid
    
    valid_ecl=['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    ecl=dic['ecl']
    if ecl not in valid_ecl:
        valid=False
        return valid
    
    pid=dic['pid']
    if len(pid)!=9 or any([f not in '0123456789' for f in dic['pid']]):
        valid=False
    return valid
```

```python
validity=[]
all_fields_passports=[]
for passport in all_passports:
    valid_fields=True
    for i in fields:
        if i not in passport.keys():
            valid_fields= False
            break
    if valid_fields == True:
        validity.append(valid_or_not(passport))
```
